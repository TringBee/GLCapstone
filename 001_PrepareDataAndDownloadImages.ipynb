GETTING ALL IMAGES WITH THEIR BOX CORDINATES

```
box1 = pd.read_excel('/content/drive/My Drive/CapStone/google-landmarks-dataset/boxes_split1.xlsx')
box2 = pd.read_excel('/content/drive/My Drive/CapStone/google-landmarks-dataset/boxes_split2.xlsx')
all_box_info = box1.append(box2)
```

```
all_box_info.columns = ['id','xmin','ymin','xmax','ymax']
all_box_info.head()
```

GETTING ALL THE TRAIN DATA AS IN GOOGLE DATASET

```
init_train = pd.read_csv('/content/drive/My Drive/CapStone/google-landmarks-dataset/train.csv')
init_train.shape
```

```
init_train = init_train[init_train['landmark_id'] != 'None'] 
init_train.shape
```

TAKING ONLY THOSE TRAIN IMAGES WITH BOX CORDINATES

```
TrainWithAllInfo = pd.merge(left=all_box_info, right=init_train)
```

```
TrainWithAllInfo.head()
```

LOADING LABELS OF LANDMARK_IDS

```
lmap = pd.read_csv('/content/drive/My Drive/CapStone/google-landmarks-dataset/landmark_map.csv')
lmap.columns=['landmark_id','landmark_name']
lmap.head()
```

```
#Converting landmark_id to int. This is needed to make a join below
TrainWithAllInfo = TrainWithAllInfo.astype({"landmark_id": int}) 
TrainWithAllInfo.dtypes
```

APPENDING LANDMARK LABELS TO DATA SET

```
TrainSuperset = pd.merge(left=TrainWithAllInfo, right=lmap)
```

```
TrainSuperset.shape
```

SUBSELECTION OF IMAGES THAT CAN BE ACTULLY DOWNLOADED

```
QualitySuperset = TrainSuperset.head(200)
QualitySuperset['downloadable'] = ''
```

ACTUAL DOWNLOAD OF IMAGES & BOOK KEEPING

```
from PIL import Image
import requests
import urllib.request
from PIL import Image
import matplotlib.pyplot as plt
c = 0
Impath = '/content/drive/My Drive/CapStone/google-landmarks-dataset/AllImages/'
for index, row in QualitySuperset.iterrows():
    c = c + 1
    try:
         im = Image.open(urllib.request.urlopen(row['url']))
         #newsize = (128, 128)
         #im = im.resize(newsize)
         imagename = "%s_%s.jpg"%(row['landmark_id'] , row['id'] )
         
         im = im.save(Impath+imagename)
         print(c, ' ', Impath+imagename)
         QualitySuperset.at[index,'downloadable'] = 'Y'         
    except:
         pass
```

DISCARDING UNDOWNLOADBALE RECORDS

```
 QualitySuperset = QualitySuperset[QualitySuperset['downloadable'] =='Y']
```

SELECTING ONLY THOSE RECORDS WHO HAVE MORE THAN 3 IMAGES PER LANDMARK

```
TrainMultiImages =  QualitySuperset.groupby('landmark_id').filter(lambda group: len(group) > 3)
```

SPLITTING INTO TRAIN AND TEST. TEST SHALL HAVE ONE IMAGE PER LANDMARK. TRAIN WILL HAVE AT LEAST MORE THAN 2 IMAGES PER LANDMARK

```
testSet = pd.DataFrame()
for i in TrainMultiImages.landmark_id.unique():
  testSet = pd.concat([testSet, TrainMultiImages[TrainMultiImages['landmark_id'] == i].head(1)])
```

```
trainSet = TrainMultiImages[~TrainMultiImages.id.isin(testSet.id)]
```

STORING THE OUTPUT

```
trainSet['filename'] = trainSet.apply(lambda x: (str(x.landmark_id)+'_'+str(x.id)+'.jpg'), axis=1)
```

```
testSet['filename'] = testSet.apply(lambda x: (str(x.landmark_id)+'_'+str(x.id)+'.jpg'), axis=1)
```

```
trainSet.to_csv('/content/drive/My Drive/CapStone/google-landmarks-dataset/AllImages/train.csv')
testSet.to_csv('/content/drive/My Drive/CapStone/google-landmarks-dataset/AllImages/test.csv')
```
