```
def show_sample(inp):
  from PIL import Image
  import requests
  import urllib.request
  from PIL import Image
  import cv2
  import matplotlib.pyplot as plt
  import numpy as np
  i = np.random.randint(inp.shape[0]) 
  xmin, ymin, xmax, ymax, label, id, landmark_id = train.loc[i, ['xmin', 'ymin', 'xmax', 'ymax','landmark_name', 'id', 'landmark_id']]
  filename = '/content/drive/My Drive/CapStone/google-landmarks-dataset/AllImages/' + str(landmark_id) +'_'+ str(id) +'.jpg'
  im = cv2.imread(filename)

  height, width, channels = im.shape
  cv2.rectangle(im, (int(xmin*width), int(ymin*height)), (int(xmax*width), int(ymax*height)), (0,255,0), 2)
  cv2.putText(im,label,(int(xmin*width), int(ymin*height+50)),cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2)
  #Convert BGR format (used by opencv to RGB format used by matplotlib)
  im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)


  plt.figure(figsize=(15,10))
  plt.imshow(im)
  plt.show()
```

```
show_sample(test)
```
